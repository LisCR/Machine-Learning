```python
 # My first project in Pythom sobre machine learning area
 # Master in Managment & Organization of knowledge - PPGGOC - ECI UFMG
 # Semester 2017-2
 # Discipline Representations distribuided of text and modeling of topics(TGI853)
 # Professor Doctor Renat Rocha
 # Student Liliane Chaves de Resende
 #
 # Objectives of project:
 #
 #  1-Representações textuais para aprendizado de máquinas. 
 #    Sumarização automática de texto
 #      objetivo: criar um algoritmo para ler um texto e resumí-lo
 #      Principais bibliotecas utilizadas: NLTK e BeatifulSoap
 #      Técnica utilizada: modelo baseado em regras, que consiste
 #                         de extrair as frases mais importantes do texto
 #                         para criar o resumo. Por meio de técnicas de LPN para:
 #                          a) encontrar as palavras mais importantes do texto
 #                          b) criar pontuação nas frases mais importantes
 #                          c) quanto maior a pontuação mais importante é a sentença
 #                          d) Pegar as n frases mais importantes para criar o resumo
 #                          e) para definir a frase mais importante será encontrado
 #                             as palavras que mais se repetem no texto e, em seguida,
 #                             criar uma distribuição de frequência
# Configurações necessárias
import nltk
# importando funções Request e urlopen de urlib.request
import urllib3
from urllib.request import Request, urlopen    
    

link = Request('http://www.nucleodeoncologia.com.br/br/p/128/textos-medicos.aspx',headers={'User-Agent': 'Mozilla/5.0'})

# variável página armazena informações retornadas em link
pagina = urlopen(link).read().decode('utf-8', 'ignore')
    
# utilizando BeautifuSoap para garimpar o site
from bs4 import BeautifulSoup
soup = BeautifulSoup(pagina, "lxml")
    
# somente busca informações da tag id
texto = soup.find(id="controleMostraPagina_ctl00_ctrlConteudo7").text

# PNL tratando o texto extraído - funções de NLTK
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize

# Dividindo o texto em sentenças
sentencas = sent_tokenize(texto)
palavras = word_tokenize(texto.lower())

# Retirando as stopwords (palavras sem relevância no texto)
from nltk.corpus import stopwords
from string import punctuation

# não permite elementos repetidos, ajustando o idioma, retirando pontuações
stopwords = set(stopwords.words('portuguese') + list(punctuation))
palavras_sem_stopwords = [palavra for palavra in palavras if palavra not in stopwords]
    
# criando distribuição de frequência - funções de NLTK
from nltk.probability import FreqDist
frequencia = FreqDist(palavras_sem_stopwords)
    
# separando sentenças importantes
from collections import defaultdict
sentencas_importantes = defaultdict(int)
    
# populando dicionário - looping percorrendo todas as sentenças e coletando estatísticas
for i, sentenca in enumerate(sentencas):
    for palavra in word_tokenize(sentenca.lower()):
        if palavra in frequencia:
            sentencas_importantes[i] += frequencia[palavra]    

# Selecionando sentenças no dicionário - funções nlargest de heapq
from heapq import nlargest

# definidos 5 sentenças mais importantes
idx_sentencas_importantes = nlargest(5, sentencas_importantes, sentencas_importantes.get)

# criando o resumo
for i in sorted(idx_sentencas_importantes):
     print(sentencas[i])
    
     
    
    
```

```python
# Bibliotecas necessárias
%matplotlib inline
from sklearn.datasets import load_iris

# biblioteca dividira a base de dados 2 dois
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier #KNN
import matplotlib.pyplot as plt
import numpy as np

# base de dados
iris = load_iris()
iris.keys()
```

```python
print(iris['DESCR'][:193] + '\n...')
```

```python
#types presente na base iris
iris['target_names'] 
```

```python
#descrição de cada característica
iris['feature_names']
```

```python
#medidas de cada flor da base iris
print(iris['data'].shape)
iris['data'][:20]
```

```python
# matriz numpy valores de 0 a 2. 
# Cada um para um tipo de flor (0-setosa 1-versicolor 2-virginica)
iris['target']
```

```python
# iniciando treinamento: conjunto de treino e conjunto de teste
X_train, X_test, y_train, y_test = train_test_split(iris['data'], iris['target'], random_state = 0)

print(X_train.shape)
print(X_test.shape)
```

```python
fig, ax = plt.subplots(3, 3, figsize=(15,15))

for i in range(3):
    for j in range(3):
        ax[i,j].scatter(X_train[:,j], X_train[:, i + 1], c=y_train, s=60)
        ax[i,j].set_xticks(())
        ax[i,j].set_yticks(())
        
        if i == 2:
            ax[i,j].set_xlabel(iris['feature_names'][j])
        if j == 0:
            ax[i,j].set_ylabel(iris['feature_names'][i + 1])
        if j > i:
            ax[i,j].set_visible(False)
```

```python
# criando o modelo de classificação Machine learning KNN
# variavel x_new - flor prevista
knn = KNeighborsClassifier(n_neighbors = 1)
knn.fit(X_train, y_train)
```

```python
# testando inserindo nova medida para a flor 
X_new = np.array([[6, 3,  1,  0]])
X_new.shape
```

```python
# prevendo qual a espécie - encontrou especie virgínica
# variavel prediction = valor da função knn.predict
prediction = knn.predict(X_new)
prediction
```

```python
iris['target_names'][prediction]
```

```python
#precisão do modelo
knn.score(X_test, y_test)
```
